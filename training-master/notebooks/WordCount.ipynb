```python
import findspark
findspark.init()
```

```python
from pyspark import SparkContext
sc = SparkContext("local[1]", "WordCount")
```

```python
# map vs flatMap

dataset = [ ["python", 1995], ["c++", 1980], ["java", 1996] ]   

rdd = sc.parallelize(dataset)
rdd2 = rdd.map (lambda lang: lang)

def show(w):
    print (w)
    return w
    
rdd3 = rdd2.map (lambda w: show(w))

print(rdd3.collect()) 
```

```python
# flattening a list
dataset = [ ["python", 1995], ["c++", 1980], ["java", 1996] ] 

rdd = sc.parallelize(dataset)
rdd2 = rdd.flatMap (lambda lang: lang)

def show(w):
    print (w)
    return w
    
rdd3 = rdd2.map (lambda w: show(w))


rdd3.collect()
```

```python
# in Linux/Ssh
# upload a file into hadoop directory
# cd ~
# wget https://raw.githubusercontent.com/brunoklein99/deep-learning-notes/master/shakespeare.txt
# hdfs dfs -copyFromLocal ./shakespeare.txt /input



fileRdd = sc.textFile("hdfs://192.168.93.128:9000/input/shakespeare.txt")  
# find the unique word count
# return an list ['AL', 'Alabama']
wordsRdd = fileRdd.map (lambda line: line.lower().split(" "))\
                  .flatMap(lambda li: li)   # return flatted element AL, Alabama, ...
    
# create keyed Rdd (key, value) (AL, 1), (Alabama, 1)
wordsPairRdd = wordsRdd.map (lambda word: (word, 1) ) # keyed rdd with key and value

# aggregate or reduce
# reduceByKey, it collect all the words, find the count of it
# acc is accumulator variable/arg, value start from 0, 0 + 1, 0 + 1 + 1
# occ occurance, the value 1 in the tuple (AL, 1)
wordCounts = wordsPairRdd.reduceByKey(lambda acc, occ: acc + occ)

# AL 1
# Alabama 1
print(wordCounts.collect())
```
