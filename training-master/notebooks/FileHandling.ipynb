```python
states_file = open("c:\\data\\all_us_states.csv")
header = states_file.readline() # read header
lines = states_file.readlines() # read rest of the lines excluding header
print ("header ", header)

# print(lines)

# create a class called State, with two attributes code, name, initialize them with __init__, __str__

class State:
    def __init__(self, code, name):
        self.code = code
        self.name = name
        
    def __str__(self):
        return f'State(code={self.code},name={self.name})'

    def __repr__(self):
        return f'State(code={self.code},name={self.name})'

states = []

for line in lines:
    line = line.strip() # remove white space, \n \t
    print (line)
    # split the string by , => output should be a list with two elements, STATE CODE, State name
    (code, name) = line.split(",")
    states.append(State(code, name))
    # for each state, create a new State object initialized with state code and name
    # add the state object to states list using append
states_file.close()

for state in states:
    print(state)
```

```python
# read a file with finally block, to close the file
try: 
    states_file = open("c:\\data\\all_us_states.csv")
    for line in states_file:
        print(line)
except:
    print("File error")
finally:
    states_file.close()
```

```python
# final and best way to handle the files using with statement
# with statement shall the close the automatically when the block exited
with open("c:\\data\\all_us_states.csv") as states_file:
    for line in states_file:
        print(line)
        
# no need to close the file, with statement will close the fiel automatically when it exit the block
```

```python
# open a file in write mode, and write content and close the file with block
# if the file not there, it will create one
# if the file already there, it will overwrite
with open("c:\\data\\employees.csv", "w") as emp_file:
    emp_file.write("name,age,gender\n")
    emp_file.write("joe,30,m\n")
    emp_file.write("mary,26,f\n")
```

```python
# open file in append mode, the content is appended, instead of overwrite
with open("c:\\data\\employees2.csv", "a") as emp_file:
    emp_file.write("joe,30,m\n")
    emp_file.write("mary,26,f\n")
```
