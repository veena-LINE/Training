### Kafka Consumer with custom JSON DeSerializer
Make sure Zookeeper + Kafka Broker are running, and `statewise_earning` topic is available.

```python
from kafka import KafkaConsumer

print("Consumer running.....")
# To consume latest messages and auto-commit offsets
consumer = KafkaConsumer('statewise_earning',
                         bootstrap_servers=['192.168.93.128:9092'])

# read messages from consumer
for message in consumer:
    """
    message value and key are raw bytes -- decode if necessary!
    e.g., for unicode: `message.value.decode('utf-8')`
    """
    """ p=partition o=offset k=key v=value """
    print(f"p={message.partition}: o={message.offset}:  k={message.key}: v={message.value.decode('utf-8')}")
#     print ("partition %d: offset %d: key=%s value=%s" % ( message.partition,
#                                           message.offset, message.key,
#                                           message.value))
```

```python
import findspark
findspark.init()
```

```python
from pyspark.conf import SparkConf

conf = SparkConf()
conf.setMaster("local").setAppName("SparkStream:KafkaConsume")


from pyspark.sql import SparkSession
ss = SparkSession.builder.config(conf=conf).getOrCreate()
```

```python
kafkaDf = (
    ss.readStream.format("kafka")
    .option("kafka.bootstrap.servers", "192.168.93.128:9092")
    .option("subscribe", "statewise_earning")
    .load()
)
```

```python
kafkaDf.printSchema()
```

```python
"""
I need only 1 thing: value
DeSerialize just the byte-stream value as only this is needed for further use.
"""

# ordersJsonRawDf = kafkaDf.selectExpr("timestamp", "CAST(value AS STRING)")
ordersJsonRawDf = kafkaDf.selectExpr("CAST(value AS STRING)")
ordersJsonRawDf.printSchema() # we get only value as string
```

```python
import json
import pyspark.sql.functions as F
from pyspark.sql.types import StructField, StructType, StringType, DoubleType


"""
JSON is an object, Spark DF needs a schema
""" 
schema = StructType(
        [
            StructField("state", StringType(), True),
            StructField("Amount", DoubleType(), True)
        ]
)


"""
Replace just a JSON-string with a JSON-object-with-schema
Now, value is a single column, it contains a struct
"""
jsonDf = ordersJsonRawDf.withColumn("value", F.from_json("value", schema))


"""
Extract JSON object elements as individual columns
"""
# ordersDf = jsonDf.select("timestamp", F.col("value.*"))
ordersDf = jsonDf.select(F.col("value.*"))
```
